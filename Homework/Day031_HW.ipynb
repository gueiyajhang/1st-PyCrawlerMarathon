### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
# 蝦皮搜尋結果的商品
r = requests.get('https://shopee.tw/api/v2/search_items/?by=relevancy&keyword=shm%20%E5%93%A5%E5%90%89%E6%8B%89&limit=50&newest=0&order=desc&page_type=search&version=2')
response = r.text

response
```

```python
import requests
headers = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://shopee.tw/api/v2/search_items/?by=relevancy&keyword=shm%20%E5%93%A5%E5%90%89%E6%8B%89&limit=50&newest=0&order=desc&page_type=search&version=2',headers=headers)
response = r.text

response
```
