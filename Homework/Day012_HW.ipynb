### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find('em').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df = pd.DataFrame(columns=['category', 'date', 'title'])
for d in soup.find(class_="part_list_2").find_all('h3'):
    df = df.append({'category':d.find('em').text, 'date':d.find(class_="date").text, 'title':d.find_all('a')[-1].text}, ignore_index=True )
print(df)
```

### ③ 哪一個類別的文章最多

```python
# 列出所有類別
list_category = df.category.unique()
list_num = []
for category in list_category:
    list_num.append({'category':category, 'num':df[df.category == category].count()[0]})
        
# 找出最多文章的類別
max_item = max(list_num, key=lambda x:x['num'])
print(max_item)
```
